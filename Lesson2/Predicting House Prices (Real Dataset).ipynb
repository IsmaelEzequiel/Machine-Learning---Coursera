# Fire up Pandas, Matplotlib and Start Use XD

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('home_data.csv')
sales = pd.DataFrame(data)
```

# Exploring the data for house sales

```python
sales.plot(kind='scatter', x='sqft_living', y='price')
```

# Creating simple prediction 

```python
# Using numpy for load my data and getting 80% of the data to use how train
mask = np.random.rand(len(sales)) < 0.8
```

```python
train = sales[mask]
```

```python
# the ~ is to get the rest of the datas
test = sales[~mask]
```

```python
plt.scatter(sales['sqft_living'], sales['price'])
plt.xlabel('Sqft House')
plt.ylabel('Price')
plt.title('House Analysis')
plt.show()
```

```python
test.plot(kind='scatter', x='sqft_living', y='price') # Train_Error
```

```python
train.plot(kind='scatter', x='sqft_living', y='price') # Train_Data for pradiction
```
