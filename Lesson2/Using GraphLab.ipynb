```python
import graphlab as lb
```

```python
sales = lb.SFrame('home_data.gl/')
```

```python
lb.canvas.set_target('ipynb')
```

```python
sales.show(view='Scatter Plot', x='sqft_living', y='price')
```

```python
train_data, train_error = sales.random_split(.8, seed=0)
```

# Build a linear regression

```python
sqft_model = lb.linear_regression.create(train_data, target='price', features=['sqft_living'])
```
